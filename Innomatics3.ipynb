```python
# Employee Performance Bonus Eligibility

# Input dictionary
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Highest score
max_score = max(employees.values())

# All employees with the highest score
top_performers = [name for name, score in employees.items() if score == max_score]

# Display result
print("Top Performers Eligible for Bonus:", ", ".join(top_performers), f"(Score: {max_score})")
```

```python
# Search Query Keyword Analysis

import string

# Input query
query = "Buy mobile phone buy phone online"

# Convert to lowercase
query = query.lower()

# Remove punctuation
query = query.translate(str.maketrans("", "", string.punctuation))

# Split into words
words = query.split()

# Count frequency
freq = {}
for w in words:
    freq[w] = freq.get(w, 0) + 1

# Filter words
result = {k: v for k, v in freq.items() if v > 1}

# Display result
print(result)
```

```python
# Sensor Data Validation

# Input list of sensor
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# Identify even readings
valid_readings = [(i, val) for i, val in enumerate(sensor_readings) if val % 2 == 0]

# Display result
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
# Email Domain Usage Analysis

# Input list of emails
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Extract domains
domains = [email.split("@")[1] for email in emails]

# Count frequency
domain_count = {}
for d in domains:
    domain_count[d] = domain_count.get(d, 0) + 1

# Calculate percentage
total = len(emails)
domain_percentage = {d: (count / total) * 100 for d, count in domain_count.items()}

# Display results
for d, pct in domain_percentage.items():
    print(f"{d}: {int(pct)}%")
```

```python
# Sales Spike Detection

# Input sales data
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average
avg_sales = sum(sales) / len(sales)

# Detect spikes
spikes = [(day+1, val) for day, val in enumerate(sales) if val > avg_sales * 1.3]

# Display results
print("Average Sales:", avg_sales)
print("Days with Sales Spikes:")
for day, val in spikes:
    print(f"Day {day}: {val}")
```

```python
# Duplicate User ID Detection

# Input list of user IDs
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Count frequency of each user ID
id_count = {}
for uid in user_ids:
    id_count[uid] = id_count.get(uid, 0) + 1

# Identify duplicates
duplicates = {uid: count for uid, count in id_count.items() if count > 1}

# Display results
print("Duplicate User IDs:")
for uid, count in duplicates.items():
    print(f"{uid} → {count} times")
```
